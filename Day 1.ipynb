```python
filename = 'input.txt'
```

```python
with open(filename, 'r') as data:
    x = []
    y = []
    for line in data:
        p = line.split('   ')
        x.append(int(p[0]))
        y.append(int(p[1]))
```

```python
x_arr = np.array(x)
y_arr = np.array(y)
```

```python
# Sort the two arrays to get the smallest numbers
x_arr_sort = np.sort(x_arr)
y_arr_sort = np.sort(y_arr)
```

```python
#For each value get distance between x and y
xy_dist = y_arr_sort - x_arr_sort
```

```python
#Get the absolute of the sum of the distance 
abs(xy_dist).sum()
```

```python
sim_score = []
```

```python
for i in x_arr_sort:
    if len(y_arr_sort[y_arr_sort == i]) == 0:
        sim_score.append(0)
    else:
        sim_score_x = len(y_arr_sort[y_arr_sort == i]) * i
        sim_score.append(sim_score_x)
```
