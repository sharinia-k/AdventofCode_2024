```python
# Find the correct order that the list should be in 
def re_order_list(list1):
    for ind_test, t in enumerate(list1):
        for i in rules:
        #First number should be before the second number 

            if i[1] == t:
                #print(i, t)
                #Get the index of 27 and check the index of the other number, if one should be before the other than raise a flag
                if i[0] in list1: #Check if the first number is in the list
                    ind = list1.index(i[0])
                    #print(ind)
                    if ind > ind_test: #If the position of the first number is after the second number, then it's wrong and needs to be re-ordered
                            list1.insert(ind_test, list1.pop(ind)) #Insert the first number in the place of the original number
                    else:
                        
                        continue
    return list1    
```
